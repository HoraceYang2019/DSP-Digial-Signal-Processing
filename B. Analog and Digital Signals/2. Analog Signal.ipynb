## <font color='blue'>2-1 Signal Representation</font>

<div>

</div>

### 類比訊號(Analog Signals)可以定義為：「隨著時間改變的連續訊號」，因此也稱為連續時間訊號(Continuous-Time Signals)。

\begin{equation}
\begin{split}
x(t) &= A\cos (\omega t + \phi) \\
     &= A\cos (2\pi ft + \phi) \\       
\end{split}  \tag{2.1}
\end{equation} 

其中，A稱為振幅(Amplitude)，$\omega$稱為角頻率(Angular Frequency)，f稱為頻率(Frequency)，$\phi$ 稱為相位移(Phase Shift)；且 
$\omega = 2 \pi f$。

## <font color='blue'>2-2 Sinusoids</font>

### Expmple 1: 
若弦波 $x(t)=Acos(2\pi ft + \phi)$ 的參數為：
　振幅：A = 1, 5, 10
　頻率：f = 1, 2, 5 (Hz)
　相位移： $\phi$ = 0, $\frac{\pi}{4}$, $\frac{\pi}{2}$ (Radians)
請比較弦波的差異，並說明其特性。

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace( 0, 1, 1000, endpoint = False )	# 定義時間陣列
f = 5
phi = np.pi/2
x = 5*np.sin( 2 * np.pi * f * t + phi)					# 產生弦波
#print(t)
plt.plot( t, x )								# 繪圖
plt.xlabel( 't (second)' )
plt.ylabel( 'Amplitude' )
plt.grid()
plt.show( )
```

## 2-2-1 Signal Addition

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace( 0, 1, 1000, endpoint = False )			# 定義時間陣列
x1 = 3 * np.cos( 2 * np.pi * 10 * t + np.pi / 4 )		# 第一個弦波
x2 = 3 * np.cos( 2 * np.pi * 10 * t - 3*np.pi / 4 )	# 第二個弦波
x3 = x1 + x2											# 弦波相加

plt.plot( t, x1, '--', label = 'x1(t)' )				# 繪圖
plt.plot( t, x2, '--', label = 'x2(t)' )
plt.plot( t, x3, '-', label = 'x3(t)' )

plt.legend( loc = 'upper right' )
plt.xlabel( 't (second)' )
plt.ylabel( 'Amplitude' )
plt.axis( [ 0, 1, -6, 6 ] )

plt.show( )
```

## <font color='blue'>2-2 Complex Numbers</font>

### 複數(Complex Numbers)可以定義為：
\begin{equation}
\begin{split}
z= a+bj  \\       
\end{split}  \tag{2.2}
\end{equation} 

其中，$a$稱為實部(Real Part)、$b$稱為虛部(Imaginary Part)； $j=\sqrt{-1}$  為虛數單位。

<div>

</div>

```python
import numpy as np

z = 3 + 1j*np.sqrt(3)							# 定義複數
magnitude = abs( z )				# 計算強度(Magnitude)
theta = np.angle( z ) * 180 / np.pi	# 計算相位角(Phase Angle)

print( "z =", z )
print( "Magnitude =", magnitude )
print( "Phase Angle =", theta )
```

### 歐拉公式
\begin{equation}
\begin{split}
e^{jx} = \cos x + j \sin x \\       
\end{split}  \tag{2.3}
\end{equation} 

## 相量(Phasor)
可以定義為：
$X=Ae^{j\phi}$
其中，A稱為振幅(Amplitude)，$\phi$ 稱為相位移(Phase Shift)。

## Phasor Addition (相量相加)
若有N個弦波，角頻率均為$\omega_0$，振幅與相位移分別為$A_k$與$\phi_k$，k = 1,…, N，則：
\begin{equation}
\begin{split}
\sum_{k=1}^n a_k \cos (\omega_0 t + \phi_k) = A \cos (\omega_o t + \phi)\\       
\end{split}  \tag{2.4}
\end{equation} 

稱為相量加法規則(Phasor Addition Rule)。

### 複數的四則運算如下：
$\begin{align}
(a+bj) + (c+dj) = (a+c) + (b+d)j  \tag{2.5}\\
(a+bj) - (c+dj) = (a-c) + (b-d)j  \tag{2.6}\\
(a+bj)(c+dj) = (ac-bd) + (bc+ad)j  \tag{2.7}\\
\frac{a+bj}{c+dj} = \frac{(ac+bd)+(bc+-ad)j}{c^2+d^2}  \tag{2.8}
\end{align}$

### Expmple 2: 
若 $x_1(t)=3cos(2\pi (10)t + \frac{\pi}{4})$\
　 $x_2(t)=4cos(2\pi (10)t - \frac{3\pi}{4})$\
What are the amplitude and phase of $x_3(t)=x_1(t)+x_2(t)$?

\begin{equation}
\begin{split}
x_3(t) &= 3cos(2\pi (10)t + \frac{\pi}{4}) + 4cos(2\pi (10)t - \frac{3\pi}{4})\\  
       &= \frac{3}{2}(e^{20 \pi t + \frac{\pi}{4} } + e^{-j 20\pi t - \frac{\pi}{4}}) + \frac{4}{2}(e^{20 \pi t - \frac{3\pi}{4}} + e^{-j 20\pi t +\frac{3\pi}{4}})
\end{split}
\end{equation}

```python
import numpy as np

A1 = 3
A2 = 4
phasor1 = complex( A1 * np.cos( np.pi / 4 ), A1 * np.sin( np.pi / 4 ) )
phasor2 = complex( A2 * np.cos( -3 * np.pi / 4 ), A2 * np.sin( -3 * np.pi / 4 ) )
phasor = phasor1 + phasor2

A = abs( phasor )					
phi = np.angle( phasor ) * 180 / np.pi

print( "Phasor1 =", phasor1 )
print( "Phasor2 =", phasor2 )
print( "Phasor =", phasor )
print( "Amplitude =", A )
print( "Phase Angle =", phi )
```

## <font color='blue'>2-4 Sinusoidal waves at various frequency</font>

Fourier Analysis	–	any composite signal can be represented as a combination of simple sine waves 
with different  frequencies, phases, and amplitudes.
\begin{equation}
s(t)=A_1 \sin (2\pi f_1 t + \phi_1)+ A_2 \sin (2\pi f_2 t + \phi_2)+...  \tag{2.9}\\       
\end{equation} 

A periodic composite signal (period=T, freq. = $f_0$=1/T) can be represented 
as a sum of simple sines and  cosines known as the Fourier series:

\begin{equation}
s(t)=\frac {A_0}{2} + \sum_{n=1}^\infty [A_n \cos (2\pi n f_0 t)+ B_n \sin (2\pi n f_0 t)]  \tag{2.10}\\       
\end{equation} 

Let $\omega = \frac {2\pi}{T} = 2\pi f$

so we have
\begin{equation}
A_n = \frac{2}{T} \int_0^T s(t)\cos (n\omega_0 t) \,dx, n=0,1,2,...    \tag{2.11}  
\end{equation}

\begin{equation}
B_n = \frac{2}{T} \int_0^T s(t)\sin (n\omega_0 t) \,dx, n=0,1,2,... \tag{2.12}       
\end{equation}

#### How to generate a wave at multiple frequencies with random lengths?
\begin{equation}
s(t) = A \sum_{i=0}^{3} (\sin (2\pi f_0 t) + sin (2 \pi f_1 t)) \tag{2.14}\\       
\end{equation} 

```python
from IPython.display import Audio
import numpy as np
import random

framerate = 48000
data = []

key = [(2 ** num) for num in range(0, 4)]
for i in range(0, 4):
    r = random.random()
    t = np.linspace(0, key[i]*(int(i*r+1)),framerate *(int(i*r+1)))
    data.append(np.sin(2 * np.pi * 440 * t) + \
          np.sin(2 * np.pi * 349.228 * t))  #double
output = np.concatenate(data[:])
Audio(output, rate = framerate)
```
