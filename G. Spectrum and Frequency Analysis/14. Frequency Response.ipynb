## <font color='blue'>Introduction to Frequency Response</font>
Frequency Response (頻率響應) is the response of input signals at specific ranges of frequencies.<br>
The response includes the values of magnitude (強度) and phase angle (相位角) and is represented in spectrum to indicate the operation characteristics.<br>
For a LTI (or DSP) system, after inputting signal $x[n]$, the output signal $y[n]$ can be derived by using the impulse response $h[n]$ (also named filter). 

<div>

</div>

If there exists $y[n]=h[n]*x[n]$, where * is the convolution operation. Then, we have 
\begin{equation}
y[n]=\sum_{k=-\infty}^{\infty} h[k] \cdot x[n-k]
\end{equation}

<br>Let $x[n]=e^{j \omega n}$, then
\begin{equation}
y[n]=\sum_{k=-\infty}^{\infty} h[k] \cdot e^{j \omega (n-k)} = (\sum_{k=-\infty}^{\infty} h[k] \cdot e^{j \omega k}) e^{j \omega n}
\end{equation}

<br >Assume $H(e^{j\omega})=\sum_{k=-\infty}^{\infty} h[k] \cdot e^{j \omega k}$, and $y[n]$ can be rewrite as 
\begin{equation}
y[n]=H(e^{j\omega}) e^{j \omega n}
\end{equation}

<br>Here, $H(e^{j\omega})$ is the frequency response of the DSP system, it also is the result of discrete-time fourier transform of $h[n]$.

Note:<br>
>  $|H(e^{j\omega})|$ is named magnitude.<br>
>  arg$\{H(e^{j\omega}) \}$ is named phase angle. <br>
>  The system gain is defined $G(\omega)=20 \log_{10} | H(e^{j\omega})|$ (dB) 

## <font color='blue'>Filter Classification</font>
>低通濾波器(Lowpass Filter): <br>
    Allow the portion of low frequencies of a signal to pass through a system, but inhibit the frequencies higher than the cutoff frequency $f_c$.<br>

>高通濾波器(Highpass Filter) <br>
    Allow the portion of high frequencies of a signal to pass through a system, but inhibit the frequencies lower than the cutoff frequency $f_c$.<br>

>帶通濾波器(Bandpass Filter) <br>
    Only allow the portion of frequencies of a signal between $f_1 - f_2$ to pass through a system, but inhibit others.<br>

>帶阻濾波器(Bandstop Filter) <br>
    Only inhibit the portion of frequencies of a signal between $f_1 - f_2$ to pass through a system, but allow others to pass through.<br>

>陷波濾波器(Notch Filter) <br>
    Only inhibit a specific frequency $f_n$ of a signal to pass through a system, but allow others to pass through.<br>

>梳狀濾波器(Comb Filter) <br>
    Inhibit multiple frequences of a signal to pass through a system, but allow others to pass through.<br>

>全通濾波器(All-Pass Filter) <br> 
    Shift phase of some frequencies of a signal.<br>

### <font color='blue'> Examples of frequency response</font>
### <font color='green'>Ideal Lowpass filter (transfer fucntion) </font>
The transfer function of an ideal lowpass filter can be defined as
\begin{equation}
    H_{LP} (e^{j \omega})=
    \begin{cases}
      1, & 0 \leq \omega \leq \omega_c \\
      0, & \text{otherwise}
    \end{cases}
\end{equation}
where $\omega_c = 2\pi f_c$, $f_c$ is the cutoff frequency.

### <font color='green'>Ideal highpass filter </font>
The transfer function of an ideal highpass filter can be defined as
\begin{equation}
    H_{HP} (e^{j \omega})=
    \begin{cases}
      1, & \omega_c < \omega \leq \pi \\
      0, & \text{otherwise}
    \end{cases}
\end{equation}
That is $H_{HP}(e^{j\omega}) = 1- H_{LP}(e^{j\omega})$.

### <font color='green'>Ideal bandpass filter </font>
The transfer function of an ideal bandpass filter can be defined as
\begin{equation}
    H_{BP} (e^{j \omega})=
    \begin{cases}
      1, & \omega_1 \leq \omega \leq \omega_2 \\
      0, & \text{otherwise}
    \end{cases}
\end{equation}
where $\omega_1$ and $\omega_2$ are the cutoff frequencies.

### <font color='green'>Ideal bandstop filter </font>
The transfer function of an ideal bandstop filter can be defined as
\begin{equation}
    H_{BS} (e^{j \omega})=
    \begin{cases}
      1, & \omega < \omega_1 \text { or } \omega > \omega_2 \\
      0, & \text{otherwise}
    \end{cases}
\end{equation}

### <font color='green'>Ideal Lowpass filter (frequency response) </font>
According to the inverse discrete-time Fourier transform, the frequency response of the lowpass filter is
\begin{equation}
\begin{split}
h_{LP}[n] &= \frac{1}{2\pi}\int_{-\infty}^{\infty} H_{LP}(e^{j \omega})e^{j \omega n} d\omega \\
          &= \frac{1}{2\pi}\int_{-\omega_c}^{\omega_c} e^{j \omega n} d\omega  \\       
          &= \frac{1}{2\pi} [\frac{1}{jn} e^{j \omega n}]_{-\omega_c}^{\omega_c} \\
          &= \frac{1}{2\pi} [\frac{e^{j \omega_c n}}{jn} - \frac{e^{-j \omega_c n}}{jn}] \\
          &= \frac{\sin(\omega_c n)}{\pi n}, -\infty < n < \infty 
\end{split}
\end{equation} 
But for $n=0$, $h_{LP}[0]$ can be derived based on the L'Hopital' s rule as follows.
\begin{equation}
\begin{split}
h_{LP}[n] &= \lim_{n \to 0} \frac{\sin(\omega_c n)}{\pi n} \\
          &= \lim_{n \to 0} \frac{\frac{d}{dn} \sin (\omega_c n)}{\frac{d}{dn}(\pi n)} 
        = \lim_{n \to 0} \frac{\cos (\omega_c n) \cdot \omega_c}{\pi} = \frac{\omega_c}{\pi}  \\       
\end{split}
\end{equation}

### Ideal Lowpass Impulse Response

```python
import numpy as np
import matplotlib.pyplot as plt

filter_size = eval( input( "Please enter the filter size (in odd): " ) )
filter_half = int( filter_size / 2 )
wc = np.pi / 2

na = np.arange( -filter_half, filter_half + 1 )	# 
h = np.zeros( filter_size )						# 
for n in na:
	if n == 0: 
		h[n+filter_half] = 0.5
	else:
		h[n+filter_half] = np.sin( wc * n ) / ( np.pi * n )

plt.stem( na, h )
plt.xlabel( 'n' )
plt.ylabel( 'h[n]' )

plt.show( )
```

### Ideal Lowpass Frequency Response

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

filter_size = eval( input( "Please enter the filter size: " ) )
filter_half = int( filter_size / 2 )
wc = np.pi / 4

na = np.arange( -filter_half, filter_half + 1 )	# 
h = np.zeros( filter_size )						# or signal.firwin(80, 0.5, window=('kaiser', 8))
for n in na:
	if n == 0: 
		h[n+filter_half] = 0.5
	else:
		h[n+filter_half] = np.sin( wc * n ) / ( np.pi * n )

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.freqz.html#scipy.signal.freqz
# scipy.signal.freqz(b, a=1, worN=512, whole=False, plot=None, fs=6.283185307179586, include_nyquist=False)
w, H = signal.freqz( h )
mag = abs( H )

plt.plot( w, mag )
plt.xlabel( r'$\omega$' )
plt.ylabel( 'Magnitude' )

plt.show( )
```

### <font color='green'>Average filter (frequency response) </font>
An average filter can be defined as 
\begin{equation}
h[n]=\frac{1}{M}\{1, 1,...,1\}, n=0,1,...,M-1
\end{equation}
where M is the filter size.

Based on the Euler's formula,
\begin{equation}
e^{j \omega} = \cos(\omega) + j\sin(\omega); \quad  \cos(\omega)=\frac{e^{j\omega}+e^{-j\omega}}{2}; 
\quad  \sin(\omega)=\frac{e^{j\omega}-e^{-j\omega}}{2j}
\end{equation}

According to the inverse discrete-time Fourier transform, the frequency response of the average filter can be derived as follows.
\begin{equation}
\begin{split}
H(e^{j \omega}) &= \sum_{n={-\infty}}^{\infty} h[n]e^{-jn \omega} 
                = \frac{1}{M} \sum_{n=0}^{M-1} e^{-jn \omega } \\
                &= \frac{1}{M} (\sum_{n=0}^{\infty} e^{-jn \omega} - \sum_{n=M}^{\infty} e^{-jn \omega}) 
                = \frac{1}{M} (\sum_{n=0}^{\infty} e^{-jn \omega})(1 - e^{-jM \omega}) \\
                &= \frac{1}{M} \frac{1-e^{-jM \omega}}{1-e^{-j \omega}} \\
\end{split}
\end{equation} 

Because
\begin{equation}
1- e^{-j \omega} = e^{\frac{-j\omega}{2}}\left( e^{\frac{j\omega}{2}} - e^{\frac{-j\omega}{2}} \right) 
                = e^{\frac{-j\omega}{2}}\left( 2j \sin(\frac{\omega}{2}) \right)
                = 2j \sin (\omega /2) e^{-j\omega /2}
\end{equation}

Also we have,
\begin{equation}
1- e^{-jM \omega} = 2j \sin (M\omega /2) e^{-Mj\omega /2}
\end{equation}

Hence,
\begin{equation}
\begin{split}
H(e^{j \omega}) &= \frac{1}{M} \frac{1-e^{-jM \omega}}{1-e^{-j \omega}} \\
                &= \frac{1}{M} \frac{\sin(M \omega/2)}{\sin(\omega/2)} e^{-j (M-1)\omega/2}
\end{split}
\end{equation} 
<br>
Then, the magnitude of the frequency response is <br>
\begin{equation}
|H(e^{j \omega})| = \left|\frac{\sin(M \omega/2)}{\sin(\omega/2)} \right|
\end{equation}

### <font color='green'>FIR filter (frequency response) </font>
A FIR lowpass filter can be defined as 
\begin{equation}
H_0(z) = \frac{1}{2} (1+ z^{-1})
\end{equation}

And, a FIR highpass filter can be defined as 
\begin{equation}
H_1(z) = \frac{1}{2} (1- z^{-1})
\end{equation}

Question: What are the frequency responses of the FIR filters?  

<div>

</div>

According to the frequency response of the lowpass filter, the maximum and minimum of the magnitudes are as follows.<br>
\begin{equation}
|H_0(e^{j0})|=1 \text{ and } |H_0(e^{j\pi})|=0
\end{equation}

For the cutoff frequency $\omega =\omega_0$, the magnitude is 
\begin{equation}
|H_0(e^{j\omega_c})| = \frac{1}{\sqrt{2}}|H_0(e^{j0})|
\end{equation}

When represent it in dB,
\begin{equation}
\begin{split}
20 \log_{10} |H_0(e^{j\omega_c})| &= 20 \log_{10} \left(\frac{1}{\sqrt{2}}|H_0(e^{j0})| \right) \\
                                  &= 20 \log_{10}|H_0(e^{j0})| - 20 \log_{10}\sqrt{2} = 0-3.0103 \approx -3 \text{dB}
\end{split}
\end{equation}
Hence, $\omega_c$ is called 3dB cutoff frequency.

$20 \log_{10}(\frac{1}{10}) = ? $ dB, $20 \log_{10}(\frac{1}{100}) = ? $ dB,

### Frequency Response of simple FIR filter 

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

b = np.array( [ 0.5, 0.5] )	# 
w, H0 = signal.freqz( b )
H0 = abs( H0 )

b = np.array( [ 0.5, 0.5, 0.1,0.1, 0.1] )	# 
w, H1 = signal.freqz( b )
H1 = abs( H1 )

plt.figure( 1 )
plt.plot( w, H0 )
plt.xlabel( r'$\omega$' )
plt.ylabel( 'Magnitude' )
plt.grid()

plt.figure( 2 )
plt.plot( w, H1 )
plt.xlabel( r'$\omega$' )
plt.ylabel( 'Magnitude' )
plt.grid()
plt.show( )
```

### <font color='green'>IIR filter (frequency response) </font>
An IIR lowpass filter can be defined as 
\begin{equation}
H_0(z) = \frac{1-\alpha}{2} \left( \frac{1+z^{-1}}{1-\alpha z^{-1}} \right)
\end{equation}

An IIR highpass filter can be defined as 
\begin{equation}
H_1(z) = \frac{1+\alpha}{2} \left( \frac{1-z^{-1}}{1-\alpha z^{-1}} \right)
\end{equation}

where $|\alpha|<1$ is a stable IIR Filter.

Question: What are the frequency responses of the IIR filters, if $\alpha =0.2, 0.4, 0.6, 0.8$?  

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

alpha = eval( input( "Please enter alpha (< 1): " ) )

b = np.array( [ 1 - alpha, 1 - alpha ] )
a = np.array( [ 2, -2 * alpha ] )

w, H = signal.freqz( b, a )
H0 = abs( H )

b = np.array( [ 1 + alpha, - ( 1 + alpha ) ] )
a = np.array( [ 2, -2 * alpha ] )
w, H = signal.freqz( b, a )
H1 = abs( H )

plt.figure( 1 )
plt.plot( w, H0 )
plt.xlabel( r'$\omega$' )
plt.ylabel( 'Magnitude' )

plt.figure( 2 )
plt.plot( w, H1 )
plt.xlabel( r'$\omega$' )
plt.ylabel( 'Magnitude' )

plt.show( )
```

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

b = np.array( [ 0.5, 0.5] )	# 
w, H0 = signal.freqz( b )  # FIR
H0 = abs( H0 )

alpha = 0.8
b = np.array( [ 1 - alpha, 1 - alpha ] )
a = np.array( [ 2, -2 * alpha ] )

w, H1 = signal.freqz( b, a )  # IIR
H1 = abs( H1 )

sigma = 3
filter_size = int( 6 * sigma + 1 )				#
gauss = signal.gaussian( filter_size, sigma )	# 
sum = np.sum( gauss )                   		# 
gauss = gauss / sum

w, H2 = signal.freqz( gauss )
H2 = abs( H2 )
plt.figure( 1 )
plt.plot(w, H0, label = 'FIR' )
plt.plot( w, H1, label = 'IIR')
plt.plot( w, H2, label = 'Gaussian')
plt.xlabel( r'$\omega$' )
plt.ylabel( 'Magnitude' )
plt.title('Low pass filters')
plt.legend();  plt.grid()
plt.show()
```

### <font color='green'>IIR badpass filter (frequency response) </font>
An IIR bandpass filter can be defined as 
\begin{equation}
H_{BP}(z) = \frac{1-\alpha}{2} \frac{1-z^{-2}}{1-\beta(1+\alpha)z^{-1}+\alpha z^{-2}}
\end{equation}

where $|\alpha|<1$ and  $|\beta|<1$ is a stable IIR Filter.

Question: What are the frequency responses of the IIR bandpass filters, if $\alpha =0.2, 0.4, 0.8$ and $\beta=0.5$?  

<div>

</div>

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

alpha = 0.2
beta = 0.5
b = np.array( [ 1 - alpha, 0, -( 1 - alpha ) ] )
a = np.array( [ 2, -2 * beta * ( 1 + alpha ), 2 * alpha ] )
w, H = signal.freqz( b, a )
H1 = abs( H )

alpha = 0.5
beta = 0.5
b = np.array( [ 1 - alpha, 0, -( 1 - alpha ) ] )
a = np.array( [ 2, -2 * beta * ( 1 + alpha ), 2 * alpha ] )
w, H = signal.freqz( b, a )
H2 = abs( H )

alpha = 0.8
beta = 0.5
b = np.array( [ 1 - alpha, 0, -( 1 - alpha ) ] )
a = np.array( [ 2, -2 * beta * ( 1 + alpha ), 2 * alpha ] )
w, H = signal.freqz( b, a )
H3 = abs( H )

plt.plot( w, H1, '-', label = r'$\alpha$ = 0.2' )
plt.plot( w, H2, '--', label = r'$\alpha$ = 0.5' )
plt.plot( w, H3, '-.', label = r'$\alpha$ = 0.8' )

plt.legend( loc = 'upper right' )
plt.xlabel( r'$\omega$' )
plt.ylabel( 'Magnitude' )
plt.grid()
plt.show( )
```
